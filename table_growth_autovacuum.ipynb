```python
import pandas as pd

df = pd.read_csv('tpcc_table_growth.csv')
df = df[['ds', 'public.customer']].rename(columns={"public.customer": "y"})
df['ds'] = pd.to_datetime(df['ds'], unit='s')
df2 = pd.read_csv('tpcc_table_growth_no_autovacuum.csv')
df2 = df2[['ds', 'public.customer']].rename(columns={"public.customer": "y"})
df2['ds'] = pd.to_datetime(df2['ds'], unit='s')
```

## Autovacuum enabled

```python
df.plot(x='ds', y='y')
```

```python
df2.plot(x='ds', y='y')
```
